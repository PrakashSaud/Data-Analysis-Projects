NumPy is the **engine** that powers Python‚Äôs data ecosystem.

```python
import numpy as np
```

### üìå Task 2: 1D Arrays ‚Äì Representing Simple Data

A **1D array** is just a list of numbers, but NumPy makes it faster and more powerful.  
Imagine tracking **daily hotel bookings** over a week.  
Each number represents the number of rooms booked for that day.  

This transforms into a **vector**, the simplest building block of numerical data.

```python
bookings = np.array([45, 52, 48, 60, 72, 50, 65])  # daily hotel bookings
print("Shape:", bookings.shape)
print("Dimensions:", bookings.ndim)
print("Bookings on 3rd day:", bookings[2])
```

### üìå Task 3: 2D Arrays ‚Äì Organizing Data into Tables

Now imagine recording bookings for **two different hotels** over the same 7 days.  
Each row represents a hotel, each column represents a day of the week.  

This is a **matrix** (2D array), which is the foundation for **data tables**.

```python
hotel_bookings = np.array([
    [45, 52, 48, 60, 72, 50, 65],  # Hotel A
    [38, 41, 55, 62, 70, 68, 75]   # Hotel B
])

print("Dimensions:", hotel_bookings.ndim)
print("Shape:", hotel_bookings.shape)
print("Bookings of Hotel B on Day 3:", hotel_bookings[1, 2])
print("Bookings for Hotel A:", hotel_bookings[0, :])
```

### üìå Task 4: N-D Arrays ‚Äì Tracking Data Over Multiple Dimensions

What if we want to analyze **three hotels** across 7 days and split it further into **two booking types**: 
- Direct bookings
- Online bookings  

Now we need a **3D array** (a tensor).  

This structure is exactly how complex data is stored: 
- Dimensions = Hotels, Days, Booking Types  
- Useful in **deep learning, image analysis, and forecasting models**.

```python
mystery_bookings = np.array([
    [[30, 15, 28, 32, 40, 22, 20],   # Hotel A (Direct)
     [15, 37, 20, 28, 32, 28, 45]],  # Hotel A (Online)

    [[25, 20, 30, 35, 42, 26, 28],   # Hotel B (Direct)
     [13, 21, 25, 27, 28, 42, 47]],  # Hotel B (Online)

    [[18, 28, 32, 34, 36, 22, 29],   # Hotel C (Direct)
     [20, 25, 28, 32, 34, 35, 40]]   # Hotel C (Online)
])

print("Dimensions:", mystery_bookings.ndim)
print("Shape:", mystery_bookings.shape)
print("Bookings for Hotel C (Online) on Day 7:", mystery_bookings[2, 1, 6])
print("All bookings for Hotel B on Day 3:", mystery_bookings[1, :, 2])
```

### üìå Task 5: Broadcasting ‚Äì Scaling Data Effortlessly

Hotel management decides to increase all room prices by **$20 per booking**.  

Instead of looping through every value, NumPy **broadcasts** the addition across the entire array.  
This saves time and makes operations much faster ‚Äì a crucial feature when working with **large datasets**.

```python
prices = np.array([[100, 120, 130],
                   [150, 160, 170]])  # base prices

new_prices = prices + 20  # broadcasting
print("Updated Prices:\n", new_prices)
```

### üìå Task 6: Linear Algebra ‚Äì The Math Behind Predictions

Linear algebra is at the core of machine learning and optimization.  
For hotels, it can help solve problems like:
- Predicting **future bookings** based on past data
- Optimizing **room allocation** between different categories  

With NumPy, we can easily perform:
- Matrix multiplication  
- Transposition  
- Determinants (used in optimization problems)

```python
A = np.array([[2, 3], [4, 5]])  # past demand
B = np.array([[1, 2], [3, 4]])  # pricing impact

print("Matrix Multiplication:\n", np.dot(A, B))
print("Transpose of A:\n", A.T)
print("Determinant of A:", np.linalg.det(A))
```

### üìå Task 7: Generating Data for Visualization

Visualizations are crucial for storytelling.  

With NumPy, we can quickly generate data points.  
For example, let‚Äôs simulate **seasonal booking trends** using a sine wave (more bookings in summer, fewer in winter).  

```python
import matplotlib.pyplot as plt

days = np.linspace(0, 365, 365)   # one year
seasonal_trend = 50 + 30 * np.sin(2 * np.pi * days / 365)  # simulate seasonal bookings

plt.figure(figsize=(10,4))
plt.plot(days, seasonal_trend, color="blue")
plt.title("üìà Seasonal Booking Trends Over a Year")
plt.xlabel("Day of Year")
plt.ylabel("Expected Bookings")
plt.show()
```

### üèÅ Challenge 1: Generating Sequential Data with `.arange()`

In data science, we often need **sequences of numbers** ‚Äì for IDs, time intervals, or simulations.  
Instead of typing numbers manually, NumPy‚Äôs `.arange()` creates them automatically.  

Example: Imagine generating **room numbers from 10 to 29** in a hotel block üè®.  
This simple trick saves time and scales easily.

```python
import numpy as np

a = np.arange(10, 30)
print("Generated array:\n", a)
```

### ‚úÇÔ∏è Challenge 2: Extracting Useful Subsets with Slicing

Hotel managers rarely need **all data at once**.  
Instead, they want:
- Last few days of booking numbers  
- A small middle segment for trend checks  
- Excluding the earliest records  
- Only **even-numbered rooms**  

NumPy slicing lets us **zoom into specific parts** of an array instantly üîç.

```python
print("Last 3 values:", a[-3:])
print("4th to 6th values:", a[3:6])
print("Values after first 12:", a[12:])
print("Only even values:", a[::2])
```

### üîÑ Challenge 3: Reversing Data Order

Sometimes, analysts need to **look back in time** ‚è≥.  
For example, reviewing bookings from **latest to earliest**.  

NumPy makes reversing order effortless:
- Either with slicing `[::-1]`  
- Or with `.flip()`  

This saves time when exploring historical data.

```python
print("Reversed array (slice):", a[::-1])
print("Reversed array (flip):", np.flip(a))
```

### üìç Challenge 4: Detecting Meaningful Data

Real-world data often has **zeros (missing or unused values)**.  
For example, `[6, 0, 9, 0, 0, 5, 0]` could represent **guest reviews** where 0 = no review.  

We want to find the **indices of non-zero values**, i.e., the positions where guests left reviews.  
This allows managers to **filter out noise** and focus on meaningful entries ‚úÖ.

```python
b = np.array([6, 0, 9, 0, 0, 5, 0])
nz_indices = np.nonzero(b)
print("Indices of non-zero values:", nz_indices)
```

### üìè Challenge 6: Generating Evenly Spaced Values

Sometimes we don‚Äôt want integers but **evenly spaced values** across a range.  
For example, dividing **discount percentages** from 0% to 100% into 9 equal steps.  

`.linspace()` makes this easy by **splitting ranges into equal intervals**.

```python
x = np.linspace(0, 100, num=9)
print("Evenly spaced values (0 to 100):\n", x)
```

### üìä Challenge 7: Visualizing Trends with Generated Data

Let‚Äôs generate values between -3 and 3 using `.linspace()` for a variable `y`.  
By pairing it with our `x` values, we can **plot a simple linear trend**.  

This is similar to plotting:
- Seasonal trends  
- Customer satisfaction scores  
- Price sensitivity curves üìà

```python
import matplotlib.pyplot as plt

y = np.linspace(-3, 3, num=9)

plt.plot(x, y, marker="o", color="blue")
plt.title("Generated Relationship Between x and y")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.show()
```

### üé® Challenge 8: Creating Images from Random Noise

Every image is just a **matrix of numbers**.
